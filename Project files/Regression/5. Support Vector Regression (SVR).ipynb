# Support Vector Regression Model

## Data Pre-Processing

### Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the Data Set

```python
ds = pd.read_csv('ifood_df.csv')
ds.info()
ds
```

#### education_Master and education_2n Cycle are equivalent things, so we merge them into one column:

```python
sum_cols = ds['education_2n Cycle'] + ds['education_Master']
ds['education_Master'] = sum_cols
ds.drop('education_2n Cycle',axis=1,inplace=True)
ds.info()
```

#### Based on the data of customers, we want to predict the total amount spent by them, so we keep the relevant columns only

```python
X=ds[ds.columns[~ds.columns.isin(['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts',
                                  'MntGoldProds','Z_CostContact','Z_Revenue', 'MntTotal','MntRegularProds'])]]
X.info()
X
```

```python
y = ds.iloc[:, 35:36].values
y
```

### Splitting the Data Set into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=17)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

## Training the Model

### Training the Support Vector Regression Model on the Training Data Set

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train.ravel())
```

### Predicting the Test Results

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
```

## Visualising the Results

### Predicted Amount Spend vs True Amount Spend for Test Data

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Amount vs Income plot for Training Set Data

```python
plt.scatter(sc_X.inverse_transform(X_train)[:, 0:1].ravel(), sc_y.inverse_transform(y_train), c='r', label='True Value',s=25)
plt.scatter(sc_X.inverse_transform(X_train)[:, 0:1].ravel(), sc_y.inverse_transform(regressor.predict(X_train).reshape(1,-1)), c='b', label='Predicted Value',s=20)
plt.title("Total Amount Spent vs Income (Train Set)")
plt.xlabel("Income")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

### Amount vs Income Plot for Test Set Data

```python
plt.scatter(sc_X.inverse_transform(X_test)[:, 0:1].ravel(), sc_y.inverse_transform(y_test), c='r', label='True Value',s=25)
plt.scatter(sc_X.inverse_transform(X_test)[:, 0:1].ravel(), sc_y.inverse_transform(y_pred), c='b', label='Predicted Value',s=20)
plt.title("Total Amount Spent vs Income (Test Set)")
plt.xlabel("Income")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

### Amount vs Recency for Train Set

```python
plt.scatter(sc_X.inverse_transform(X_train)[:, 3:4].ravel(), sc_y.inverse_transform(y_train), c='r', label='True Value',s=25)
plt.scatter(sc_X.inverse_transform(X_train)[:, 3:4].ravel(), sc_y.inverse_transform(regressor.predict(X_train).reshape(-1,1)), c='b', label='Predicted Value',s=20)
plt.title("Recency vs Income (Train Set)")
plt.xlabel("Recency")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

### Amount vs Recency for Test Set

```python
plt.scatter(sc_X.inverse_transform(X_test)[:, 3:4].ravel(), sc_y.inverse_transform(y_test), c='r', label='True Value',s=25)
plt.scatter(sc_X.inverse_transform(X_test)[:, 3:4].ravel(), sc_y.inverse_transform(y_pred), c='b', label='Predicted Value',s=20)
plt.title("Recency vs Income (Test Set)")
plt.xlabel("Recency")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

## Performance of the Model

### Evaluating the Model Performance

```python
import statsmodels.api as sm
result = sm.OLS(y_test, y_pred).fit()
print('R-squared Value:',result.rsquared,'\nR-squared Adjusted Value', result.rsquared_adj)
```
